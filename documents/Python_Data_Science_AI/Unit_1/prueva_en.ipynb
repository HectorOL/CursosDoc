### 4. Conditionals

```python
# Case 1: No secondary condition
age = 17
if age >= 18:
  print('You can enter!')
```

```python
# Case 2: 1 secondary condition
age = 15
if age >= 18:
  print('You can enter!')
else:
  diff = 18-age
  if diff == 1:
    print(f'You cannot enter, come back in 1 year!')
  else:
    print(f'You cannot enter, come back in {18-age} years!')
```

```python
# Challenge Exercise 03 - Level 2
# Assume we have a database with the name and age of several people
# Each record is a list;
# The first element is the name, the second is the age

# Challenge: Create a mechanism to select only adults (of legal age)

# Database
data = [
    ['Jonatan',26],
    ['Joshua',22],
    ['Paola',17],
    ['Ingrid',23],
    ['Julian',7]
]

# Algorithm
adults = []
for person in data:
  if person[1] >= 18:
    adults.append(person[0])
print(adults)
```
