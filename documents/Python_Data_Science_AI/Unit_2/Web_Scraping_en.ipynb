### Get HTML from a real page

```python
import requests
```

```python
url = 'https://www.blankhall.com/'
```

```python
web = requests.get(url)
```

```python
soup = BeautifulSoup(web.text, 'html.parser')
```

```python
soup.h1.text
```

```python
soup.p.text
```

```python
about = soup.find(id='about')
```

```python
for skill in about.find_all('li'):
  print(skill.text)
```

```python
skills = {}
for s in about.find_all('li'):
  first_split = s.text.split(':')
  second_split = first_split[1].split(',')
  skills[first_split[0]] = [sk.strip() for sk in second_split]
print(skills)
```

```python
skills['Mathematics']
```

## Example: free market

```python
product = "ipad pro"
```

```python
def transform_product_name(product):
  first_shape = product.split(' ')
  first_shape = '-'.join(first_shape)

  second_shape = product.split(' ')
  second_shape = '%20'.join(second_shape)
  return first_shape, second_shape
```

```python
def get_product_url(product):
  insert1, insert2 = transform_product_name(product)
  url = f'https://listado.mercadolibre.com.mx/{insert1}?sb=all_mercadolibre# D[A:{insert2}]'
  return url
```

```python
session = requests.Session()
session.headers.update({
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/115.0.0.0 Safari/537.36"
    ),
    "Accept-Language": "es-MX,es;q=0.9,en;q=0.8",
    "Accept": (
        "text/html,application/xhtml+xml,application/xml;"
        "q=0.9,image/webp,image/apng,*/*;q=0.8"
    ),
    "Referer": "https://www.mercadolibre.com.mx/",
    "Connection": "keep-alive",
})
```

```python
web = session.get(get_product_url(product))
```

```python
web
```

```python
soup = BeautifulSoup(web.text, 'html.parser')
```

```python
soup.text
```

```python
main = soup.find(id='root-app')
```

```python
products = main.find(class_="ui-search-layout ui-search-layout--stack")
```

```python
for prod in products.find_all('li'):
  prod_ = prod.find(class_ = 'poly-component__title')
  price = prod.find(class_ = 'poly-price__current')
  display = [price.text,prod_.text]
  print(display)
```
