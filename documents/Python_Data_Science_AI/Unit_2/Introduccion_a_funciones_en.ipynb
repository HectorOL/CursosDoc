Default parameters in Python

They are parameters that will be optional for the function, the user can put them or not, however of antemeno it is chosen what that value is

```python
def repetir_texto(texto,veces=2):   # We are declaring a default parameter that is sometimes = 2
  return texto*veces


repetir_texto("Hey")
```

```python
repetir_texto("Hola mundo",10)
```

$M_n = M_{n-1} + M_{n-1}*r + 5000$

$M_n = M_{n-1}(1+r) + 5000$

```python
import matplotlib.pyplot as plt

costo_casa = 10000000
ahorro = 50000
meses = costo_casa/ahorro
meses = int(meses)
anios = meses/12
rend = (0.086/12)

def ahorro_mensual(ahorro,meses,rend):

  ahorro_inversion = [ahorro]

  ahorro_sin_inversion = [ahorro]


  for i in range(1,meses):
    ahorro_inversion.append(ahorro_inversion[i-1]*(1+rend)+ahorro)
    ahorro_sin_inversion.append(ahorro_sin_inversion[i-1]+ahorro)

  plt.figure()

  plt.plot(ahorro_inversion,label="Con inversión")
  plt.plot(ahorro_sin_inversion,label="Sin inversión")
  plt.xlabel("Meses")
  plt.ylabel("Monto en pesos")
  plt.title("Ahorro mensual de una persona invirtiendo en CETES")

  plt.legend()

  plt.show()

```

```python
ahorro = 3000
meses = 100
rend = 0.15/12

ahorro_mensual(ahorro,meses,rend)
```
