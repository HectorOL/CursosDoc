Which of the classes, has a higher percentage of survival?

```python
p1 = df.query("Pclass == 1")
p2 = df.query("Pclass == 2")
p3 = df.query("Pclass == 3")
```

```python
surv1 = p1["Survived"].mean()
surv2 = p2["Survived"].mean()
surv3 = p3["Survived"].mean()


print(f"El porcentaje de pasajeros que sobrevivieron en la clase 1 fue {surv1*100}%")
print(f"El porcentaje de pasajeros que sobrevivieron en la clase 2 fue {surv2*100}%")
print(f"El porcentaje de pasajeros que sobrevivieron en la clase 3 fue {surv3*100}%")
```
