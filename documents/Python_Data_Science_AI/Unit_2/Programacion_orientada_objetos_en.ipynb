Attributes: characteristics of each object
Methods: Functions defined within a class, which can operate with attributes

```python
import random

class Pokemon:

  def __init__(self, name, health, attack_power):
    self.name = name
    self.health = health
    self.attack_power = attack_power

  # ## We create a method that allows us to attack each other
  def attack(self,pokemon2):
    damage = self.attack_power + random.randint(-5,5)
    pokemon2.health -= damage
    print(f"{self.name} ha realizado {damage} puntos de daño a {pokemon2.name} ahora tiene {pokemon2.health} puntos de vida")
```

```python
pika = Pokemon("Pikachu",100,16)
charmander = Pokemon("Charmander",100,15)


```

```python
class Battle:
  "Gestiona un combate entre dos pokemon"
  def __init__(self, pokemon1, pokemon2):
    self.pokemon1 = pokemon1
    self.pokemon2 = pokemon2

  def start_battle(self):
    while self.pokemon1.health > 0 and self.pokemon2.health > 0:
      var_trans = random.randint(0,10)
      if var_trans == 0:
        self.pokemon1.attack_power *= 2
        print(self.pokemon1.name, "se ha transformado")
      elif var_trans == 1:
        self.pokemon2.attack_power *= 2
        print(self.pokemon2.name, "se ha transformado")
      self.pokemon1.attack(self.pokemon2)
      self.pokemon2.attack(self.pokemon1)


    if self.pokemon1.health > 0:
      print(f"{self.pokemon1.name} ha ganado")
    else:
      print(f"{self.pokemon2.name} ha ganado")
```

```python
turtle = Pokemon("Turtle",100,10)
raichu = Pokemon("Raichu",100,15)


batalla1 = Battle(turtle,raichu)
batalla1.start_battle()
```

Encapsulation: controls access to attributes.

By convention for an attribute to be private, they are put __ two low scripts before the name of the attributes for example

Self .__ Password

```python
class User:
  def __init__(self, username, password):
    self.username = username
    self.__password = password     # We are defining that password is a private attribute

  def get_password(self):
    res = input("Introduce tu contraseña de super admin: ")
    if res == "189":
      print("Correct password")
      return self.__password
    else:
      print("Incorrect password")

```

Inheritance: allows a daughter class to receive attributes that were already associated with a father class

```python
class Sajayin:
  def __init__(self,nombre,attack,power):
    self.name = nombre
    self.attack_power = attack
    self.power = power

  def attack(self, personaje2):
    damage = self.attack_power + random.randint(-5,5)
    personaje2.health -= damage

  def hablar(self):
    print("Hi, I'm a Sajayin")

```

```python
class SuperSajayin(Sajayin):
  def __init__(self,name,attack):
    super().__init__(nombre = name, attack= attack , power = 10000)
```

```python
goku = SuperSajayin("Goku",1000)
```

```python
class Pikachu(Pokemon):
  def __init__(self):
    super().__init__(name = "Pikachu", health = 100, attack_power = 16)

  def hablar(self):
    print("Long")

class Charmander(Pokemon):
  def __init__(self):
    super().__init__(name = "Charmander", health = 100, attack_power = 15)

  def hablar(self):
    print("Char Char")
```
