```python
# ## Conditionals in Python

# ## The conditionals are structures that will allow us to change the flow of the program
# ## It is always good in a program to make decisions.

edad = int(input("Dame tu edad"))

if edad<18:  # if + condition +:
  print("You can't go to my bar")
  print("You have 18 years old")

else:
  print("Welcome to the Python bar")
```

There is a slightly more complicated structure, which is elif

```python
edad = 15

if edad<18:   # For Python all this is a block
  print("Eres un joven")  # Python leaves that block once any of the conditions is met
elif edad>=18 and edad<60:
  print("You are an adult")
else:
  print("You are an old man")
```

```python
edad = 10

if edad<18:
  print("Eres un joven")  # Here ends the first block
if edad>=18 and edad<60:
  print("You are an adult")
else:
  print("Eres un anciano") # Here ends the second block
```

While Conditional Structure

```python
i = 1

while i < 11:   # While + condition +:
  print(i)
  i += 1    # + = is a shortcut to write i = i + 1
  # Amount: we must update the variable on which we are asking

print("End of the program")
```

```python
# ## Guess the secret number

# ## We give a hidden number and the user must guess the number
# ## If the hidden number is larger, we say that the secret number is greater
# ## If the hidden number is smaller, we say that the secret number is smaller


import random as rd

num_secreto = rd.randint(0,10)

contador = 1

intento = int(input("Adivina el número secreto"))

while intento != num_secreto:
  print("You have not won")
  if intento > num_secreto:
    print("The secret number is smaller")
  else:
    print("The secret number is bigger")
  intento = int(input("Adivina el número secreto"))
  contador += 1

print(f"Le atinaste en el intento {contador}")
```
