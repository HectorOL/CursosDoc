### Logical operators

```python
nombre1 = "Sofia"   # An equal is to as well as a value
nombre2 = "Sófia"

nombre1 == nombre2  # #Do equal allows us to ask a question
```

```python
a = 45
b = 23

print(a > b) # a is greater than B
print(a < b) # a is less than B
print(a >= b) # a is greater than or equal to b
print(a <= b) # a is less than or equal to b
print(a != b) # a is different than b
```

### LEXICOGRAPHIC ORDER IN PYTHON

```python
a = "Aa"   # An equal is to as well as a value
b = "Ab"

print(a > b) # a is greater than B
print(a < b) # a is less than B
print(a >= b) # a is greater than or equal to b
print(a <= b) # a is less than or equal to b
print(a != b) # a is different than b
```

```python
nombre1 = "Benito"
nombre2 = "Benita"


edad1  = 26
edad2 = 20


print(nombre1 > nombre2 and edad1 > edad2)   # The operator and returns True when all conditions are true
```
