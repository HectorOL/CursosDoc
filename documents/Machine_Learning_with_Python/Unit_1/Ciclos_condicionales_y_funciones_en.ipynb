```python
# ### smaller than 12 child
# ### Minor than 18 adult
# ## Less than 40 young adult
# ## Less than 60 adultoo
# ## greater than or equal to 60 old age

edad = input("Dime tu edad:")
edad = int(edad)

if edad < 12:
  print("He is a child")
elif edad < 18:   # Elif's syntax is exactly the same as if
  print("He is an adult")
elif edad < 40:
  print("He is a young adult")
elif edad < 60:
  print("He is an adult")
else:   # Else never has a condition is executed if and only if none of the other conditions are met
  print("Is a senior age")

print("End of the program")
```

## Cycles

* Cycle for: It will allow us to execute a block of code a certain number of times
* While cycle: It allows us to execute a block of code as long as a condition is being met


```python
for i in range(0,100):
  print(i)
```

```python
for i in range(100,2,-1):   # For the Range operator there are 3 arguments, where you start, where you finish, step size
  print(i)
```

```python
for num in range(5): # FOR + AUXILIARY VARIABLE + IN + ITERABLE OBJECT
  print(num)
```

```python
frutas = ["Manzana","Fresa","Guayaba","Sandia"]


for fruta in frutas:  # FOR + AUXILIARY VARIABLE + IN + ITERABLE OBJECT
  print(fruta)        # We leave tabulation
```

```python
ingresos = [100,300,500,1000,2000,0,100]


suma = 0

for ingreso in ingresos:
    suma = suma + ingreso

print(suma)
```

### While: It deals when we want to execute a block of code whenever a condition is met

```python
i = 1

while i < 11:
  print(i)
  i = i + 1
```
