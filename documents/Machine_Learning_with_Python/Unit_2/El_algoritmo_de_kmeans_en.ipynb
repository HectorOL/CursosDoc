```python
import pandas as pd # handle data



df = pd.read_csv("/content/iris.csv",header = None)

df.head()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Graph 1: Column 1 vs column 2
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.scatterplot(x=df[0], y=df[1], hue=df[4])
plt.title('Columna 1 vs Columna 2')
plt.xlabel('Columna 1')
plt.ylabel('Columna 2')

# Graph 2: Column 3 vs Column 4
plt.subplot(1, 2, 2)
sns.scatterplot(x=df[2], y=df[3], hue=df[4])
plt.title('Columna 3 vs Columna 4')
plt.xlabel('Columna 3')
plt.ylabel('Columna 4')

plt.tight_layout()
plt.show()
```

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Climb the data for K-Means
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means (assuming 3 clusters for iris data)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```
