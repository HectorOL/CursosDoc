### 4. Conditional

```python
# Case 1: No Secondary Condition
age = 17
if age >= 18:
  print('You can enter!')
```

```python
# Case 2: 1 Secondary condition
age = 15
if age >= 18:
  print('You can enter!')
else:
  dif = 18-age
  if dif == 1:
    print(f'No puedes entrar, vuelve en 1 año!')
  else:
    print(f'No puedes entrar, vuelve en {18-age} años!')
```

```python
# Challenge Exercise 03 - Level 2
# Assume that we have a database with the name and age of several people
# Each record is a list;
# The first element is the name, the second age

# Challenge: Create a mechanism to select only adults

# Database
data = [
    ['Jonatan',26],
    ['Joshua',22],
    ['Paola',17],
    ['Ingrid',23],
    ['Julian',7]
]

# Algorithm
token = []
for person in data:
  if person[1] >= 18:
    token.append(person[0])
print(token)
```
