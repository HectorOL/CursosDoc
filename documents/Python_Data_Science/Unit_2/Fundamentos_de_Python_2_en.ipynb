## While cycle:


We occupy it when we do not know how many times we must execute an action or do not have an iterable object of which to support ourselves

```python
# ## While is executed along with a condition and while that condition is being met
# ## that is inside while

palabra = "Hola mundo"

i = 0


while i < len(palabra):  # While + condition +:
    print(palabra[i])    # Identation
    i += 1
    print(i < len(palabra))

print("End of the program")
```

```python
# ## Program that adds numbers
# ## adds as many numbers as those of the user

suma = 0

r = "s"

while r == "s":
  num = int(input("Podrias ingresar un número "))
  suma = suma + num
  r = input("¿Deseas ingresar otro número? (s/n)")

print(f"La suma de los números es {suma}")
```

### Application

Financial application:

Compare the savings we will have if we keep our money under a mattress for 10 years.


To compare it with the savings we will have if we keep it in Cetes.


$m_0$ = 5000

$m_1$ = 5000 + 5000(0.1/12)

$m_2$ = 5000 + $m_1(0.1/12)$

$m_n$ = 5000 + $m_{n-1}(0.1/12)$ + $m_{n-1}$

$m_n$ = 5000 + $m_{n-1}(1 + 0.1/12)$

```python
ahorro_mensual = 5000

tasa_anual = 0.1


def calcular_retorno(ahorro_mensual,tasa_anual,años):
  meses = años*12  # 3 years ---> 36 months
  saldo = 0
  historial = [ahorro_mensual]  # Save how much money would have every month

  for mes in range(meses):
    ahorro_nuevo = historial[-1]*(1 + tasa_anual/12) + ahorro_mensual # I'm taking m_ {n-1}
    historial.append(ahorro_nuevo)

  return historial
```

```python
ahorro_mensual = 5000

tasa_anual = 0.1

años = 10

retorno_10 = calcular_retorno(ahorro_mensual,tasa_anual,años)
retorn_0 = calcular_retorno(ahorro_mensual,0,años)
```

```python
import matplotlib.pyplot as plt


plt.figure()
plt.title("Comparación de invertir cetes vs no invertir")
plt.xlabel("Meses")
plt.ylabel("Retorno")
plt.plot(retorno_10,label="Retorno 10 años con interes")
plt.plot(retorn_0,label="Retorno 10 años sin interes")
plt.legend()
plt.show()
```
