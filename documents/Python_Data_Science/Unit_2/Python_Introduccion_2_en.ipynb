## Conditional

```python
# ## The conditionals are structures that will allow us to change the flow of the program
# ## Always in a program we have to make decisions
# ## It is good to have a structure that allows us to do


nombre_bar = "Bar python"

edad = int(input("Cuál es tu edad: "))

if edad>=18:
  print(f"Bienvenido a mi bar {nombre_bar}")
else:
  print("You can't pass yet you need to grow a little more")
```

```python
# # Girl bar
# ## The conditionals are structures that will allow us to change the flow of the program
# ## Always in a program we have to make decisions
# ## It is good to have a structure that allows us to do


nombre_bar = "Bar python"

edad = int(input("Cuál es tu edad: "))
genero = input("Cuál es tu genero: ")

if edad>=18 and genero == "Femenino":  # IF + Condition +:
  print(f"Bienvenido a mi bar {nombre_bar}")
  print("You can pass")
elif edad>=18 and genero!="Femenino":
  print("You can only pass girls")
  print("You can't pass")
else:
  print("You can't pass yet you need to grow a little more")

print("End of the program")

```
