```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming that you already have 'df' loaded
# We create a new dataframe with only the necessary columns
df_filter = df.query("Fare<150")
df_box = df_filter[["Sex", "Fare"]].dropna()

# We graph
plt.figure(figsize=(8, 6))
sns.boxplot(x="Sex", y="Fare", data=df_box, palette="Set2")
plt.title("Boxplot de tarifas por sexo", fontsize=14)
plt.ylabel("Tarifa")
plt.xlabel("Sexo")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_filter[df_filter["Sex"] == "male"], x="Fare", color="skyblue", label="Hombres", kde=True, stat="density")
sns.histplot(data=df_filter[df_filter["Sex"] == "female"], x="Fare", color="pink", label="Mujeres", kde=True, stat="density")
plt.title("DistribuciÃ³n de tarifas por sexo", fontsize=14)
plt.xlabel("Tarifa")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

## Group by

The Groupby function allows us to work our data grouping the data according to some category, in that case, for example we could work according to the ticket class or if the person is a man or woman

Allows us to create summaries quickly and easy

```python
df[df["Age"]>20]
```

```python
df.query("Age>20")
```

```python
df.groupby("Pclass")["Fare"].mean()  # Because variable grouping, how variable I am interested, using the statistical that interests me (aggregation function)
```

### Pivot Table

It allows us to make new tables, where it is perhaps easier to handle the information

It allows to reorganize the data, applying an aggregation function to the data by rows and columns

```python
pivot = pd.pivot_table(df, values= "Age", index="Pclass", columns="Survived", aggfunc="mean")

print("Pivot table de edad media (Pclass vs Survived)")
pivot
```

### Strategies to work with null data

* Eliminate lines

```python
df["Age"].fillna(df["Age"].mean(), inplace=True)

df.head()
```

```python
def categorizar_edad(edad):
  if edad <10:
    return "niÃ±o"
  elif edad <20:
    return "joven"
  elif edad <60:
    return "adulto"
  else:
    return "anciano"
```

```python
categorizar_edad(10)
```

```python
df["Age_cat"] = df["Age"].apply(categorizar_edad)
df.head()
```

```python
df.groupby("Age_cat")["Fare"].mean()
```

```python
df.groupby("Age_cat")["Survived"].mean()
```

```python
df.groupby(["Age_cat","Sex"])["PassengerId"].count()
```
