* Class: A plane or mold that defines the characteristics or behavior of a set of objects.
* Object: It is an instance of a class

* Attributes: Attributes are variables that store and define the characteristics of my object.

* METHODS: These are functions or actions that my object can perform

```python
import random

class Personaje:
  def __init__(self,nombre,vida = 100,poder_ataque = 10):
    self.nombre = nombre
    self.vida = vida
    self.poder_ataque = poder_ataque

  def estar_vivo(self):
    if self.vida <= 0:
      print(f"{self.nombre} ha muerto")
    return self.vida > 0

  def recibir_ataque(self,ataque):
    self.vida -= ataque

  def atacar(self,otro):
    daño = random.randint(-10,10)
    daño_completo = daño + self.poder_ataque
    otro.recibir_ataque(daño_completo)
    if otro.vida <0:
      otro.vida = 0

    print(f"{self.nombre} ataca a {otro.nombre} y le hace daño {daño_completo}")
    print(f"La vida de {otro.nombre} es {otro.vida}")

```

```python
mario = Personaje("Mario")
```

```python
mario.vida
```

```python
goku = Personaje("Goku",100,10)
vegeta = Personaje("Vegeta",100,10)


goku.atacar(vegeta)

```

```python
goku.atacar(vegeta)


vegeta.vida
```

# Encapsulation


* To make an attribute private we occupy __ follow the name of the attribute
* This serves as it allows us that not anyone can access this attribute
* This is known as encapsulation

```python
class Cuenta_Banco:
  def __init__(self,titular,saldo):
    self.__titular = titular   # To make an attribute private, we occupy __ follow the name of the attribute
    self.__saldo = saldo       # This serves since it allows us that not anyone can access this attribute

  def get_saldo(self):
    respuesta = input("Podrias decir la contraseña para ver el saldo")
    if respuesta == "1234":
      return self.__saldo
    else:
      print("Incorrect password")

  def get_titular(self):
    return self.__titular

  def depositar(self,monto,otro):
    self.__saldo -= monto
    otro.__saldo += monto
    print(f"El saldo de {self.__titular} es {self.__saldo}")
    print(f"El saldo de {otro.__titular} es {otro.__saldo}")

  def retirar(self,monto):
    self.__saldo -= monto
```

# Inheritance

The inheritance allows to create new classes that reuse and extend a class behaviors.

```python
class Saiyajin(Personaje):
  def __init__(self, nombre, vida, poder_ataque, nivel_transformacion):
    super().__init__(nombre, vida,poder_ataque)
    self.nivel_transformacion = nivel_transformacion

  def transformar(self):
    self.nivel_transformacion += 1
    self.poder_ataque += 10
    print(f"{self.nombre} se ha transformado en super saiyajin fase {self.nivel_transformacion}")
    print(f"Su poder de ataque ha subido a {self.poder_ataque}")
```

```python
Goku = Saiyajin("Goku",100,10,0)

Goku.nivel_transformacion
```

```python
Goku.transformar()
```

## Polymorphism

Polymorphism allows the same method to behave in different ways according to the class that is implemented

```python
class Guerrero(Personaje):
  def atacar(self,otro):
    """
    Método atacar modificado: este guerrero tiene una técnica especial, que puede causar daño critico
    """

    daño_base = random.randint(0,10)
    daño_completo = daño_base + self.poder_ataque
    num_aux = random.randint(0,10)
    if num_aux == 1:
      print(f"{self.nombre} esta furioso y ha disparado su ataque especial")
      otro.vida = otro.vida - (daño_completo * 5)
      print(f"{self.nombre} ataca a {otro.nombre} y le hace daño {daño_completo * 5}")
    else:
      otro.vida = otro.vida - daño_completo
      print(f"{self.nombre} ataca a {otro.nombre} y le hace daño {daño_completo}")
```
